# Classifier using NLP vectorized features

Frankly, since there was a column which included text descriptions submitted by the user, I saw this as an opportunity to learn and apply Natrual Language Processing. Basic and common techniques were using sklearn's text processing sub-library. However, there was also an opportunity to gain experience in developing custom pipelines through sklearn.

The first cell simply loads the data into a dataframe while removing entries that were answered untimely. These don't have resolutions, and thus have invalid target labels.

```python
import pandas as pd
import numpy as np

#Defining what dtype to convert each column to
#numberic columns are transformed after reading in
dtype_dict = {'Product':"category",
             'Consumer consent provided?': "category",
             'Submitted via': "category",
             'Company response to consumer': "category",
             'Consumer disputed?': "category"}

#read in .csv file, dates are parsed into datetime objects. 
#The Complaint ID is Unique in every entry, so it can be used as index
df = pd.read_csv('Consumer_Complaints.csv',
                 index_col=['Complaint ID'],
                 parse_dates=["Date received","Date sent to company"],
                 dtype=dtype_dict)


dfUntimelyResponse = df[df["Company response to consumer"] == "Untimely response"]
df = df[df["Company response to consumer"] != "Untimely response"]

df.head()
```

The following cell continues filtering the data out to only entries with submitted narratives. This trimmed the row count to a third of the full dataset, but its still a decent size around 300 thousand entries

```python
df_narr = df[[ 'Consumer complaint narrative', 'Company response to consumer']].copy()
df_narr = df_narr[df_narr['Consumer complaint narrative'].notna()]
df_narr = df_narr.drop(df_narr[df_narr["Company response to consumer"].isna()].index)

print('Percent of data with narrative: ', 100 * len(df_narr)/len(df))
#note that all rows with narrative are not in progress
```

### BaseEstimator & TransormerMixin Parentclass
Below, a custom transformer was defined by creating a class that inherets from sklearn's BaseEstimate and TransformerMixin.

### Beautiful Soup
Beautiful Soup is a HTML scraping library which facilitates removing HTML and text formatting tags from text.

### Natural Language Toolkit (NLTK)
NLTK library was used to remove stopwords from the corpus. Stopwords are a list of words that supply little to no information when vectorize, such as gramtical articles like "the" and prepositional phrases. The language should be speficified when using this.

The class was define using the conventions specified on the sklearn's tutorial page for creating custom column transformers.

```python
from sklearn.base import BaseEstimator, TransformerMixin

#Create Text Preprocessor that
from bs4 import BeautifulSoup
import re
from nltk.corpus import stopwords

#to lower, removes tags, removes non-letters encrypted XXXXs and stopwords. Returns words join with spaces
class CleanTextTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        pass
    
    def fit(self, X, y = None):
        return self
    
    def cleanText(self, obj):
        bs = BeautifulSoup(obj).get_text()
        letters_only = re.sub('[^a-zA-Z]', ' ', bs)
        words_only = letters_only.lower().split()
        stops = set(stopwords.words('english'))
        words = [w for w in words_only if not w in stops]
        return (" ".join(words).replace('xxxx', '').replace('xx',''))
    
    def transform(self, X, y = None):
        X['Consumer complaint narrative'] = X['Consumer complaint narrative'].apply(self.cleanText)
        return X['Consumer complaint narrative']
    
clean_text = CleanTextTransformer()

X = df_narr.drop("Company response to consumer", axis=1)
#clean_text.transform(X)
```

### Text-frequency Inverse Document-frequency (TF-IDF) Vectorizer
A common method of transforming text into a vector of numbers (recall that Machine Learning algorithms only accept numberical data) is the TF-IDF vectorization. This creates a vector length equal to the number of unique words in the corpus being processed, or up to a cap. This vectorizer was capped at 5000 to be able to fit in memory. 

Each value in the vector represents how many times the word associated to the index appears in a text string with a linear relationship, and how strings the word appears accross the whole document with a inverse logorithmic relationship.

```python
#Create TfIDfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer(analyzer='word',
                           token_pattern=r'\w{1,}',
                            max_features=5000)
```

The text is split for training and testing. The Pipeline was built to clean the text and then vectorize it with on ".fit()" call from the pipeline. The target was transfromed into multiclass representation. 

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils


X = df_narr.drop("Company response to consumer", axis=1)
Y = df_narr["Company response to consumer"]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

#Columns to word vectorize

categorical_transformer = Pipeline(steps=[('cleantxt', clean_text),('tfidf',tf_idf)])

#building the column transformer with both transformers
preprocessor = ColumnTransformer(transformers=[('cat', categorical_transformer, ['Consumer complaint narrative'])])

#fit the preprocessor, then transform trainging and test set, assign sparse matrix to variables
preprocessor.fit(X)
encX_train = preprocessor.transform(X_train)
encX_test = preprocessor.transform(X_test)

#Creating Dummy variable for target. Will be neccessary for DNN. 
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)

dummy_y = np_utils.to_categorical(encoded_Y)

dummy_y_train = np_utils.to_categorical(encoder.transform(y_train))
dummy_y_test = np_utils.to_categorical(encoder.transform(y_test)) 
```

## Logistic Regression

Since Logistic Regression has been having the best results throughout this project, it was the choice of model to train.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

lr = LogisticRegression(n_jobs=-1, solver='saga', penalty='l1', multi_class='multinomial')
lr_para = {'C':[1.0,0.1,0.01], 
           'class_weight':[None,'balanced'],
           'max_iter':[100,125]}

#Apply grid search with above parameters specified
fitmodel = GridSearchCV(lr, lr_para,cv=5, scoring='accuracy', n_jobs=-1)
fitmodel.fit(encX_train,y_train)

#store the best fitting LogisiticRegression(), create prediciton from X_test data
bestfitLR = fitmodel.best_estimator_

bestfitLR.fit(encX_train,y_train)
y_pred = bestfitLR.predict(encX_test)
print(bestfitLR.score(encX_test,y_test))

#display the result
print(classification_report(y_test, y_pred))
```

Comparatively, the results were not too bad compared to the original mutliclassed logisitic regression that was trained in a previous notebook. It didn't beat out the best model, but it didn't under preform either.

## Bernuoulli Naive Bayes

In retrospect, this model shouldn't have been trained, since it assumes features independently affect the probability of classification, which seems like a weak assumption when words are used in conjunction to express concepts. Evidently the results were less than desirable.

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

bnb = BernoulliNB()
bnb_para = {'alpha':[1,2,10,0],
            'fit_prior': [True,False],
            'binarize': [0,0.1, 0.5]}

#Apply grid search with above parameters specified
fitmodel = GridSearchCV(bnb, bnb_para,cv=3, scoring='accuracy', n_jobs=-1, verbose=10)
fitmodel.fit(encX_train,y_train)

#store the best fitting GradientBoostingClassifier(), create prediciton from X_test data
bestfitBNB = fitmodel.best_estimator_

bestfitBNB.fit(encX_train,y_train)
y_pred = bestfitBNB.predict(encX_test)
print(bestfitBNB.score(encX_test,y_test))

#display the result
print(classification_report(y_test, y_pred))
print(bestfitBNB)
```

## DNN (not enough memory)

There was an attempt to use a Deep Neural Network to classify complaints using the text vectorization. Unfortunately, the hardware that was available did not have enough memory to being able to train the model.

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report
from keras_tqdm import TQDMNotebookCallback
import tensorflow as tf


#Build model with 2 FC layers each 50 nodes (relu activation), and 1 node output output layer (sigmoid activation)
model = Sequential()
model.add(Dense(50, kernal_initializer='uniform', input_dim=5000, activation='relu'))
model.add(Dense(50, kernal_initializer='uniform', activation='relu'))
model.add(Dense(4, kernel_initializer= 'uniform', activation = 'softmax'))

#setup earlystop callback
earlystop_callback = EarlyStopping(monitor='accuracy', min_delta=0.0001, patience=2)

#Compiled with adam optimizer, binary crossentropy loss function, accuracy metric for evaluation
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])

#train the model over 100 epochs
model.fit(encX_train, dummy_y_train, epochs=20, batch_size=2, callbacks=[TQDMNotebookCallback(), earlystop_callback], verbose=0)

#print the results
_, accuracy = model.evaluate(encX_test, dummy_y_test)
print('Accuracy: %.2f' % (accuracy*100))

#Save the trained model into an .h5 file
model.save("NPL_data-model-50-50.h5")

pred = model.predict(encX_test)

pred_h1 = tf.one_hot(tf.argmax(pred, axis = 1), depth = 4)

print(classification_report(dummy_y_test, pred_h1.eval(session=tf.compat.v1.Session()), target_names= encoder.classes_))
```
